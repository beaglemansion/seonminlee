```python
headers = {
    'Authorization' : 'Basic eWhvaEBpdGV5ZXMuY28ua3I6YkFzYkp4WGFuakNx',
    'Accept' : 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9',
    'Cookie' : 'ASPSESSIONIDSGABTBCB=BBPDBJGBOLNMEEGPADFICDOO'
    }
cnt = 0
base_url = 'https://kbase.frontarena.com'

# 2. for문으로 url마다 접속하기
for url in url_list:
    
    try:
        resp = requests.get(url, headers = headers)
        soup = BeautifulSoup(resp.content, 'html.parser')
        print('original url: ',url)
    except Exception as e:
        print(e)
        print('resp받지 못한 url : ', url)
        print('link_list로는 ',cnt,'번째 url')
    
    big_table = soup.find('table', class_='main')
    
    
    title = big_table.select('td[align = left] p.title b')[0].text.replace(':','')
    
    
    try:
        target = big_table.select('td[align = left] table a')[0]['href']
    except Exception as e:
        print(e)
        print('target을 잡지 못했습니다.\n문제의 url은: ',url,)
        print('link_list로는 ',cnt,'번째 url')
        

    

    # 3. target url 만들기
    target_url = base_url + target
    print(target_url)
    if target_url[-3:] == 'pdf':
        
        file_type = '.pdf'
        
    elif target_url[-3:] == 'chm':
        
        file_type = '.chm'
        
    else:
        print(target_url[-3:],'pdf도 chm도 아닌')
        file_type = ''
        
    # 4. pdf 다운받기
    # [FCA2299-11] Data Management_BDP Instrument Expiry
    try:
        
        file_number = target_url.split('/')[-1].replace('.pdf','') # 'FCA4435-05'
        name = '['+file_number+']'+' '+ title + file_type
        
    except Exception as e:
        print(e)
        print('file_number를 만들지 못했습니다.\n문제의 url은: ',url)
        print('link_list로는 ',cnt,'번째 url')
    
    try:
        target_url = base_url + target
        rq = urllib.request.Request(target_url, headers = headers)
        res = urllib.request.urlopen(rq).read()
        with open(name, mode="wb") as f:
            f.write(res)
            print("finish...")
            print(name)
            print(cnt,'번째 완료')
        
    except Exception as e:
        print(e)
        print('파일 다운로드를 못했습니다.\n문제의 url은: ',url)
        print('link_list로는 ',cnt,'번째 url')
    
    cnt += 1
    
```
